```python
train_data['manhatton_distance'] = abs(train_data['Horizontal_Distance_To_Hydrology']) + abs(train_data['Vertical_Distance_To_Hydrology'])
train_data['Euc_distance'] = train_data['Horizontal_Distance_To_Hydrology']**2 + train_data['Vertical_Distance_To_Hydrology']**2

train_data.Euc_distance = train_data.Euc_distance.apply(lambda x: math.sqrt(x))
```

```python
test_data['manhatton_distance'] = abs(test_data['Horizontal_Distance_To_Hydrology']) + abs(test_data['Vertical_Distance_To_Hydrology'])
test_data['Euc_distance'] = test_data['Horizontal_Distance_To_Hydrology']**2 + test_data['Vertical_Distance_To_Hydrology']**2

test_data.Euc_distance = test_data.Euc_distance.apply(lambda x: math.sqrt(x))
```
